# Factor analysis

Factor analysis algorithm is implemented in the scikit-learn package:

```python
from sklearn.decomposition import FactorAnalysis
```

To illustrate how you can train this algorithm we use boston data from sckikit-learn.

```python
boston = ds.load_boston()
X = boston.data
X.shape
```

```python
model = FactorAnalysis(n_components=2, random_state=324)
model.fit(X)
X_reduced = model.transform(X)
```

```python
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(1, 1, 1)
ax.plot(X_reduced[:,0], X_reduced[:,1], 'o')
plt.show()
```

Now we will use some toy data to make it more clear.

```python
C = np.array([[0., -0.2], [3.5, -.7]])
X = np.dot(np.random.randn(1000, 2), C)
X
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(1, 1, 1)
ax.plot(X[:,0], X[:,1], 'o')
plt.show()
```

```python
model = FactorAnalysis(n_components=1, random_state=324)
model.fit(X)
X_reduced = model.transform(X)
plt.hist(X_reduced)
```

# Principal component analysis

```python
from sklearn.decomposition import PCA
```

```python
boston = ds.load_boston()
X = boston.data
X.shape
```

```python
model = PCA(n_components=2)
model.fit(X)
X_reduced = model.transform(X)
```

```python
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(1, 1, 1)
ax.plot(X_reduced[:,0], X_reduced[:,1], 'o')
plt.show()
```
